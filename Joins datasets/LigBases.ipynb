```python
import pandas as pd

df_competitivo = pd.read_csv(r'../Jogos Seleções 2018 - 2026/jogos_tratado.csv')
# 1. Carrega a base do Ranking FIFA
df_ranking = pd.read_csv(r'../Raking Fifa 2018 - 2026/historico_fifa_ranking_2018_2026.csv')

# 2. Pega as seleções únicas diretamente da base do FIFA (sem copiar e colar)
lista_ranking_fifa = df_ranking['Team'].unique().tolist()

# 3. Agora cruzamos com a lista de seleções do nosso dataset de jogos (df_competitivo)
# Pegamos todas as seleções únicas que jogaram (casa e fora)
selecoes_jogos = pd.concat([df_competitivo['home_team'], df_competitivo['away_team']]).unique()

# 4. Verificamos QUAIS seleções dos nossos jogos NÃO estão no Ranking FIFA
selecoes_perdidas = [time for time in selecoes_jogos if time not in lista_ranking_fifa]

print(f"Temos {len(selecoes_perdidas)} seleções no dataset de jogos que não encontraram correspondência no Ranking FIFA:")
print(selecoes_perdidas)
```

```python
print("Padronizando os nomes das seleções para o padrão FIFA...")

# 1. Criar o dicionário de mapeamento ("Nome antigo" : "Nome FIFA")
dicionario_traducoes = {
    'Saint Vincent and the Grenadines': 'St Vincent and the Grenadines',
    'United States Virgin Islands': 'US Virgin Islands',
    'Saint Kitts and Nevis': 'St Kitts and Nevis',
    'Cape Verde': 'Cabo Verde',
    'Ivory Coast': "Côte d'Ivoire",
    'DR Congo': 'Congo DR',
    'Saint Lucia': 'St Lucia',
    'South Korea': 'Korea Republic',
    'Iran': 'IR Iran',
    'Kyrgyzstan': 'Kyrgyz Republic',
    'North Korea': 'Korea DPR',
    'Brunei': 'Brunei Darussalam',
    'United States': 'USA',
    'Taiwan': 'Chinese Taipei'
}

# 2. Aplicar a substituição nas duas colunas do dataset de jogos simultaneamente
df_competitivo['home_team'] = df_competitivo['home_team'].replace(dicionario_traducoes)
df_competitivo['away_team'] = df_competitivo['away_team'].replace(dicionario_traducoes)

# 3. Validar se o dicionário funcionou (rodando a mesma checagem de antes)
selecoes_jogos_atualizadas = pd.concat([df_competitivo['home_team'], df_competitivo['away_team']]).unique()
selecoes_perdidas_agora = [time for time in selecoes_jogos_atualizadas if time not in lista_ranking_fifa]

print(f"✅ Substituição concluída!")
print(f"Agora temos apenas {len(selecoes_perdidas_agora)} seleções sem correspondência (todas sem filiação à FIFA):")
print(selecoes_perdidas_agora)
```

```python
# 1. Carrega a base do Ratings EA Sports
df_rating = pd.read_csv(r'../Ratings EA Sports 2018 - 2026/selecoes_sofifa_historico.csv')

# 3. Agora cruzamos com a lista de seleções do nosso dataset de jogos (df_competitivo)
# Pegamos todas as seleções únicas que jogaram (casa e fora)
selecoes_jogos2 = df_rating['Nome da Seleção'].unique()

# 4. Verificamos QUAIS seleções dos nossos jogos NÃO estão no Rating EA Sports
selecoes_perdidas = [time for time in selecoes_jogos2 if time not in selecoes_jogos]

print(f"Temos {len(selecoes_perdidas)} seleções no dataset de jogos que não encontraram correspondência no Rating FIFA:")
print(selecoes_perdidas)
```

```python
print("--- PADRONIZANDO BASE EA SPORTS ---")

# 1. Carrega a base do Ratings EA Sports
df_rating = pd.read_csv(r'../selecoes_sofifa_historico.csv')

# 2. Criar o dicionário de mapeamento ("Nome EA Sports" : "Nome Padronizado no df_competitivo")
dicionario_ea_sports = {
    'Türkiye': 'Turkey',
    'Czechia': 'Czech Republic', # Verifique se no df_competitivo está 'Czech Republic' ou 'Czechia'
    'United States': 'USA',
    'Iran': 'IR Iran',
    'Curacao': 'Curaçao' # Verifique se no df_competitivo está com cedilha ou sem
}

# 3. Aplicar a substituição na coluna de nomes do df_rating
# Substitua 'Nome da Seleção' pelo nome exato da coluna no seu CSV, se for diferente
df_rating['Nome da Seleção'] = df_rating['Nome da Seleção'].replace(dicionario_ea_sports)

# 4. Pegamos todas as seleções únicas que jogaram (casa e fora) no dataset já padronizado
selecoes_jogos_atualizadas = pd.concat([df_competitivo['home_team'], df_competitivo['away_team']]).unique()

# 5. Lista de seleções que estão no EA Sports após a renomeação
selecoes_ea_sports = df_rating['Nome da Seleção'].unique()

# 6. Identificar quais seleções do df_competitivo NÃO estão no EA Sports
selecoes_sem_rating = [time for time in selecoes_jogos_atualizadas if time not in selecoes_ea_sports]

print(f"Temos {len(selecoes_sem_rating)} seleções no dataset de jogos que não estão no EA Sports.")

print(f"Temos {len(selecoes_sem_rating)} seleções no dataset de jogos que não estão no EA Sports.")

# 7. Incluir as seleções faltantes dentro do df_rating PARA TODAS AS DATAS
# ATENÇÃO: Substitua 'Data' pelo nome real da coluna de datas do seu df_rating
col_data_rating = 'Data'

# Pega todas as datas/tempos únicos que já existem no dataset da EA
datas_unicas = df_rating[col_data_rating].dropna().unique()

# Cria todas as combinações possíveis de (Seleção Faltante x Data)
df_faltantes = pd.MultiIndex.from_product(
    [selecoes_sem_rating, datas_unicas],
    names=['Nome da Seleção', col_data_rating]
).to_frame(index=False)

# Concatenamos com o df_rating original
df_rating = pd.concat([df_rating, df_faltantes], ignore_index=True)

# (Opcional, mas recomendado) Ordenar o dataset inteiro pela data após concatenar
df_rating = df_rating.sort_values(by=col_data_rating).reset_index(drop=True)

print(f"✅ Inclusão concluída! O df_rating agora possui {len(df_rating['Nome da Seleção'].unique())} seleções únicas em todas as datas.")
```

```python
print("Extraindo seleções únicas de cada base...")

# 1. Transformamos as listas de seleções únicas em "sets" (conjuntos) para facilitar a comparação
selecoes_competitivo = set(pd.concat([df_competitivo['home_team'], df_competitivo['away_team']]).unique())
selecoes_rating = set(df_rating['Nome da Seleção'].unique())
selecoes_ranking = set(df_ranking['Team'].unique())

print(f"\nTotal de seleções únicas:")
print(f"- Base de Jogos: {len(selecoes_competitivo)}")
print(f"- Base EA Sports: {len(selecoes_rating)}")
print(f"- Base Ranking FIFA: {len(selecoes_ranking)}")

# 2. Verificação de igualdade absoluta
if selecoes_competitivo == selecoes_rating == selecoes_ranking:
    print("\n✅ SUCESSO ABSOLUTO! Todas as 3 bases possuem exatamente as mesmas seleções.")
else:
    print("\n⚠️ Há diferenças entre as bases. Analisando os cruzamentos:")

    # 3. Cruzamento: Base de Jogos vs EA Sports
    faltam_no_ea = selecoes_competitivo - selecoes_rating
    if not faltam_no_ea:
        print("\n✅ EA Sports contem TODAS as seleções da base de jogos.")
    else:
        print(f"\n❌ {len(faltam_no_ea)} seleções nos JOGOS que NÃO estão no EA SPORTS:")
        print(sorted(list(faltam_no_ea)))

    # 4. Cruzamento: Base de Jogos vs Ranking FIFA
    faltam_no_fifa = selecoes_competitivo - selecoes_ranking
    if not faltam_no_fifa:
        print("\n✅ Ranking FIFA contem TODAS as seleções da base de jogos.")
    else:
        print(f"\n❌ {len(faltam_no_fifa)} seleções nos JOGOS que NÃO estão no RANKING FIFA (Provavelmente não filiadas):")
        print(sorted(list(faltam_no_fifa)))

    # 5. Cruzamentos inversos (Opcional - Seleções que estão no FIFA/EA mas não têm jogos registrados)
    fifa_sem_jogos = selecoes_ranking - selecoes_competitivo
    if fifa_sem_jogos:
        print(f"\nℹ️ {len(fifa_sem_jogos)} seleções no RANKING FIFA que não possuem partidas na base de jogos.")

    ea_sem_jogos = selecoes_rating - selecoes_competitivo
    if ea_sem_jogos:
        print(f"\nℹ️ {len(ea_sem_jogos)} seleções no EA SPORTS que não possuem partidas na base de jogos.")
```

```python
# ---------------------------------------------------------
# 0. PREPARAÇÃO DOS DADOS (Conversão de Datas e Ordenação)
# ---------------------------------------------------------
print("Preparando e ordenando as datas...")

# Substitua 'Date' ou 'Data' pelos nomes exatos das colunas de data em cada CSV seu
col_data_jogo = 'date'
col_data_ranking = 'rank_date' # Coloque o nome da coluna de data do df_ranking
col_data_rating = 'Data' # Coloque o nome da coluna de data do df_rating

# Converter para datetime
df_competitivo[col_data_jogo] = pd.to_datetime(df_competitivo[col_data_jogo])
df_ranking[col_data_ranking] = pd.to_datetime(df_ranking[col_data_ranking])
df_rating[col_data_rating] = pd.to_datetime(df_rating[col_data_rating])

# OBRIGATÓRIO PARA O MERGE_ASOF: Ordenar do mais antigo para o mais novo
df_competitivo = df_competitivo.sort_values(col_data_jogo)
df_ranking = df_ranking.sort_values(col_data_ranking)
df_rating = df_rating.sort_values(col_data_rating)

# ---------------------------------------------------------
# 1. PASSO 1: MERGE DOS JOGOS COM RANKING FIFA
# ---------------------------------------------------------
print("Iniciando merge temporal com Ranking FIFA...")

# 1A. Puxar o Ranking do Time da CASA
df_temp = pd.merge_asof(
    df_competitivo,
    df_ranking.add_prefix('home_'),            # Adiciona prefixo para não confundir
    left_on=col_data_jogo,                     # Data base (do jogo)
    right_on='home_' + col_data_ranking,       # Data de busca (do ranking)
    left_by='home_team',                       # Nome da seleção no df_competitivo
    right_by='home_Team',                      # Nome da seleção no df_ranking
    direction='backward'                       # Pega o último ranking ANTES do jogo
)

# 1B. Puxar o Ranking do Time de FORA
df_temp = pd.merge_asof(
    df_temp,
    df_ranking.add_prefix('away_'),
    left_on=col_data_jogo,
    right_on='away_' + col_data_ranking,
    left_by='away_team',
    right_by='away_Team',
    direction='backward'
)

# Limpando a sujeira (Removendo chaves redundantes do ranking para não pesar a RAM)
cols_pra_dropar_ranking = ['home_Team', 'away_Team', 'home_' + col_data_ranking, 'away_' + col_data_ranking]
df_temp = df_temp.drop(columns=cols_pra_dropar_ranking, errors='ignore')

# ---------------------------------------------------------
# 2. PASSO 2: MERGE DO RESULTADO COM RATINGS EA SPORTS
# ---------------------------------------------------------
print("Iniciando merge temporal com Ratings EA Sports...")

# 2A. Puxar o Rating do Time da CASA
df_final = pd.merge_asof(
    df_temp,
    df_rating.add_prefix('home_'),
    left_on=col_data_jogo,
    right_on='home_' + col_data_rating,
    left_by='home_team',
    right_by='home_Nome da Seleção',
    direction='backward'
)

# 2B. Puxar o Rating do Time de FORA
df_final = pd.merge_asof(
    df_final,
    df_rating.add_prefix('away_'),
    left_on=col_data_jogo,
    right_on='away_' + col_data_rating,
    left_by='away_team',
    right_by='away_Nome da Seleção',
    direction='backward'
)

# Limpando a sujeira (Removendo chaves redundantes do rating)
cols_pra_dropar_rating = ['home_Nome da Seleção', 'away_Nome da Seleção', 'home_' + col_data_rating, 'away_' + col_data_rating]
df_final = df_final.drop(columns=cols_pra_dropar_rating, errors='ignore')
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

print("--- INICIANDO IMPUTAÇÃO POR REGRESSÃO ---")

# 1. Definimos quem é o nosso eixo X (A base para a previsão)
# Olhando seu print anterior, a coluna de pontos do ranking é 'home_TotalPoints'
coluna_x = 'home_TotalPoints'

# 2. Definimos quem são os eixos Y (O que queremos prever)
colunas_y = ['home_GER', 'home_ATQ', 'home_MED', 'home_DEF']

# Instanciamos o modelo matemático
modelo = LinearRegression()

# Vamos fazer a previsão para cada uma das colunas (OVR, ATQ, MED, DEF)
for col_y in colunas_y:

    # 3. Separar quem TEM dado (para treinar) e quem NÃO TEM (para prever)
    # Condição: O X não pode ser nulo E o Y não pode ser nulo
    df_treino = df_final[df_final[col_y].notnull() & df_final[coluna_x].notnull()]

    # Condição: O Y é nulo (precisa prever), mas o X EXISTE
    df_prever = df_final[df_final[col_y].isnull() & df_final[coluna_x].notnull()]

    if df_prever.empty:
        continue # Se não tem o que prever nesta coluna, pula para a próxima

    # 4. Preparar os eixos X e Y para o formato que o scikit-learn exige (2D array)
    X_treino = df_treino[[coluna_x]]
    y_treino = df_treino[col_y]

    X_prever = df_prever[[coluna_x]]

    # 5. Treinar o modelo de Regressão Linear com os dados conhecidos
    modelo.fit(X_treino, y_treino)

    # 6. Fazer a previsão para as linhas vazias
    y_previsto = modelo.predict(X_prever)

    # Opcional: Arredondar para número inteiro, já que Rating do FIFA não tem vírgula
    y_previsto = np.round(y_previsto)

    # 7. Injetar as previsões de volta no DataFrame final
    df_final.loc[df_prever.index, col_y] = y_previsto

    print(f"✅ {len(y_previsto)} valores imputados para a coluna {col_y}.")

print("\nNulos após Regressão:")
print(df_final[colunas_y + [coluna_x]].isnull().sum())
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

print("--- INICIANDO IMPUTAÇÃO POR REGRESSÃO ---")

# 1. Definimos os eixos X e colunas Y para cada lado
lados = {
    'home': {
        'coluna_x': 'home_TotalPoints',
        'colunas_y': ['home_GER', 'home_ATQ', 'home_MED', 'home_DEF']
    },
    'away': {
        'coluna_x': 'away_TotalPoints',
        'colunas_y': ['away_GER', 'away_ATQ', 'away_MED', 'away_DEF']
    }
}

modelo = LinearRegression()

for lado, config in lados.items():
    print(f"\n🔄 Processando atributos do {lado.upper()}...")
    coluna_x = config['coluna_x']
    colunas_y = config['colunas_y']

    for col_y in colunas_y:

        df_treino = df_final[df_final[col_y].notnull() & df_final[coluna_x].notnull()]
        df_prever = df_final[df_final[col_y].isnull() & df_final[coluna_x].notnull()]

        if df_prever.empty:
            continue

        X_treino = df_treino[[coluna_x]]
        y_treino = df_treino[col_y]
        X_prever = df_prever[[coluna_x]]

        modelo.fit(X_treino, y_treino)
        y_previsto = np.round(modelo.predict(X_prever))

        df_final.loc[df_prever.index, col_y] = y_previsto
        print(f"  ✅ {len(y_previsto)} valores imputados para '{col_y}'.")

print("\nNulos após Regressão:")
colunas_verificar = (
    lados['home']['colunas_y'] + [lados['home']['coluna_x']] +
    lados['away']['colunas_y'] + [lados['away']['coluna_x']]
)
print(df_final[colunas_verificar].isnull().sum())
```

```python
print("--- TRATANDO NULOS RESIDUAIS (SEM PONTOS DA FIFA) ---")

# 1. Definimos as colunas que ainda precisam de tratamento
colunas_ea = ['home_GER', 'home_ATQ', 'home_MED', 'home_DEF','away_GER','away_ATQ', 'away_MED', 'away_DEF']

# 2. Encontramos o pior valor (mínimo) do seu dataset para cada uma dessas colunas
minimos_ea = df_final[colunas_ea].min()

# 3. Preenchemos os Ratings da EA com o mínimo
for col in colunas_ea:
    df_final[col] = df_final[col].fillna(minimos_ea[col])

# 4. Fazemos o mesmo para os pontos da FIFA, afinal o ML também não aceita nulo nessa coluna
minimo_pontos_fifa = df_final['home_TotalPoints'].min()
df_final['home_TotalPoints'] = df_final['home_TotalPoints'].fillna(minimo_pontos_fifa)

minimo_pontos_fifa = df_final['away_TotalPoints'].min()
df_final['away_TotalPoints'] = df_final['away_TotalPoints'].fillna(minimo_pontos_fifa)

print("✅ Imputação residual concluída!")
print("\nNulos finais na base de Casa (Home):")
print(df_final[colunas_ea + ['home_TotalPoints']].isnull().sum())
```
