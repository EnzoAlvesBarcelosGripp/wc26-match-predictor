```python
import os
import glob
from bs4 import BeautifulSoup
import csv

nome_ficheiro_csv = 'selecoes_sofifa_historico.csv'

# 1. Busca arquivos "tabela_*.html" em TODAS as subpastas a partir de onde o código está rodando
arquivos_html = glob.glob("**/tabela_*.html", recursive=True)

if len(arquivos_html) == 0:
    print("Nenhum arquivo HTML encontrado. Verifique se o script/notebook está na pasta raiz do projeto.")
else:
    print(f"Encontrados {len(arquivos_html)} arquivos para processar. Iniciando extração...\n")

    with open(nome_ficheiro_csv, mode='w', newline='', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file)

        # Cabeçalho com a coluna de Data
        writer.writerow(['Data', 'Nome da Seleção', 'GER', 'ATQ', 'MED', 'DEF'])

        total_geral = 0

        for caminho_arquivo in arquivos_html:

            # Pega só o nome do arquivo, ignorando as pastas (Ex: "tabela_06_14_2018.html")
            nome_arquivo = os.path.basename(caminho_arquivo)

            # Extrair a data do nome do arquivo
            texto_data = nome_arquivo.replace('tabela_', '').replace('.html', '')
            data_formatada = texto_data.replace('_', '/')

            # Ler o conteúdo HTML
            with open(caminho_arquivo, 'r', encoding='utf-8') as f:
                html_content = f.read()

            soup = BeautifulSoup(html_content, 'html.parser')

            tabela_body = soup.find('tbody')
            if tabela_body:
                linhas = tabela_body.find_all('tr')
            else:
                linhas = soup.find_all('tr')

            contador_arquivo = 0
            for linha in linhas:
                coluna_nome = linha.find('td', class_='s20')
                if not coluna_nome:
                    continue

                nome_selecao = coluna_nome.find('a').text.strip()
                ger = linha.find('td', attrs={'data-col': 'oa'}).text.strip()
                atq = linha.find('td', attrs={'data-col': 'at'}).text.strip()
                med = linha.find('td', attrs={'data-col': 'md'}).text.strip()
                defesa = linha.find('td', attrs={'data-col': 'df'}).text.strip()

                # Escreve a linha
                writer.writerow([data_formatada, nome_selecao, ger, atq, med, defesa])

                contador_arquivo += 1
                total_geral += 1

            print(f"✅ [{data_formatada}] Extraídas {contador_arquivo} seleções de: {caminho_arquivo}")

    print(f"\n🚀 Extração concluída com sucesso! Total de {total_geral} registros salvos no arquivo '{nome_ficheiro_csv}'.")
```
